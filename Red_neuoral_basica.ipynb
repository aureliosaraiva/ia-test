```python
import tensorflow as tf
import numpy as np
```

```python
celsies = np.array([-40, -10, 0, 8, 15, 22, 38], dtype=float)
fahrenheit = np.array([-40, 14, 32, 46, 59, 72, 100], dtype=float)
```

```python
capa = tf.keras.layers.Dense(units=1, input_shape=[1])
modelo = tf.keras.Sequential([capa])
```

```python
modelo.compile(
    optimizer=tf.keras.optimizers.Adam(0.1),
    loss='mean_squared_error'
)
```

```python
print("Comenzando entreinamiento...")
historial = modelo.fit(celsies, fahrenheit, epochs=1000, verbose=False)
print("Modelo entrenado!")
```

```python
import matplotlib.pyplot as plt
plt.xlabel("# epochs")
plt.ylabel("Magnitud de pérdida")
plt.plot(historial.history['loss'])
```

```python
print("Hahgamos una prediccíon!")
resultado = modelo.predict([300])
print("El resultado es " + str(resultado) + " fahrenheit!")
```
